```python
import kagglehub
import os
import shutil
import cv2
import matplotlib.pyplot as plt
```

```python
path = kagglehub.dataset_download("usamarana/skin-tone-classification-dataset")

print("Path to dataset files:", path)
```

```python
current_location = os.getcwd()[:-5] + "\\dataset"

if not os.path.exists(current_location):
    os.makedirs(current_location)

for item_name in os.listdir(path):
    full_item_path = os.path.join(path, item_name)
    dest_item_path = os.path.join(current_location, item_name)
    
    if os.path.isdir(full_item_path):
        shutil.move(full_item_path, dest_item_path)
        print(f"Moved folder: {full_item_path} → {dest_item_path}")
    elif os.path.isfile(full_item_path):
        shutil.move(full_item_path, dest_item_path)
        print(f"Moved file: {full_item_path} → {dest_item_path}")

print(f"dataset has been moved to: {current_location}")
```

### **Some sample images**

```python
def show_image(dataset_path: str, image_name: str, color = "Black"):
    sample_img_path = os.path.join(dataset_path, color, image_name)

    image = cv2.imread(sample_img_path)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
show_image(dataset_path, "1_1_1_20161219205055053.jpg.chip.jpg", "Black")
```
